## Library Size (Total Counts per Sample)
Check if samples have similar sequencing depth

```python
import matplotlib.pyplot as plt
import numpy as np

library_sizes = counts.sum(axis=0)

plt.figure(figsize=(10, 5))
plt.bar(library_sizes.index, library_sizes.values, color='steelblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Sample')
plt.ylabel('Total Counts')
plt.title('Library Size per Sample')
plt.tight_layout()
plt.show()

print(f"Library size range: {library_sizes.min():,.0f} - {library_sizes.max():,.0f}")
```

## Count Distribution
Log-transformed count distribution per sample

```python
# Log2 transform (add 1 to avoid log(0))
log_counts = np.log2(counts + 1)

plt.figure(figsize=(10, 5))
log_counts.boxplot()
plt.xticks(rotation=45, ha='right')
plt.xlabel('Sample')
plt.ylabel('Log2(counts + 1)')
plt.title('Distribution of Log2 Counts per Sample')
plt.tight_layout()
plt.show()
```

## Sample Correlation Heatmap
Check for outliers and batch effects

```python
import seaborn as sns

corr_matrix = log_counts.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=0.8, vmax=1)
plt.title('Sample-to-Sample Correlation (Log2 Counts)')
plt.tight_layout()
plt.show()
```

## PCA Analysis
Visualize sample clustering by condition

```python
from sklearn.decomposition import PCA

# PCA on transposed log counts (samples as rows)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(log_counts.T)

pca_df = pd.DataFrame({
    'PC1': pca_result[:, 0],
    'PC2': pca_result[:, 1],
    'sample': meta['sample'],
    'condition': meta['condition']
})

plt.figure(figsize=(8, 6))
for condition in pca_df['condition'].unique():
    subset = pca_df[pca_df['condition'] == condition]
    plt.scatter(subset['PC1'], subset['PC2'], label=condition, s=100)
    
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.title('PCA of Samples')
plt.legend()
plt.tight_layout()
plt.show()
```

## Gene Detection Rate
Number of detected genes (count > 0) per sample

```python
genes_detected = (counts > 0).sum(axis=0)

plt.figure(figsize=(10, 5))
plt.bar(genes_detected.index, genes_detected.values, color='darkgreen')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Sample')
plt.ylabel('Number of Detected Genes')
plt.title('Gene Detection Rate per Sample')
plt.tight_layout()
plt.show()

print(f"Detected genes range: {genes_detected.min():,} - {genes_detected.max():,}")
```
